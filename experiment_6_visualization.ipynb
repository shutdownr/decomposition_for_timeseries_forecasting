```python
%load_ext autoreload
%autoreload 2

import pickle

import matplotlib.pyplot as plt
import numpy as np
from config import Config

plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "Times New Roman"
```

```python
RESULTS_FOLDER = "results/experiment_6/DLinear/"
c = Config()
```

```python
evaluation_metric = "mae"
all_errors = {}
for dataset in c.dataset_names:
    with open(RESULTS_FOLDER + dataset + ".pkl", "rb") as file:
        errors = pickle.load(file)
        for name, e in errors.items():
            if name in all_errors:
                all_errors[name].append(e[evaluation_metric])
            else:
                all_errors[name] = [e[evaluation_metric]]

mean_errors = {}
matrix_errors = np.array(list(all_errors.values()))
means = np.mean(matrix_errors,axis=0)
stds = np.std(matrix_errors,axis=0)
for name, errs in all_errors.items():
    normalized_error = (errs - means) / stds
    mean_errors[name] = np.mean(normalized_error)

for decomp in ["none", "moving_avg", "fourier_bandlimited", "fourier_topk", "wavelet"]:
    filtered_dict = {key: value for key, value in mean_errors.items() if key.startswith(decomp)}
    min_error_index = np.argmin(list(filtered_dict.values()))
    min_error = list(filtered_dict.values())[min_error_index]
    best_configuration = list(filtered_dict.keys())[min_error_index]
    configs = best_configuration[len(decomp)+1+len("decomposition")+1:]
    print("Decomposition:", decomp, "Best Configuration:", configs, "Error:", min_error)

min_errors = {}
for dataset in c.dataset_names:
    with open(RESULTS_FOLDER + dataset + ".pkl", "rb") as file:
        errors = pickle.load(file)
        errs = [d[evaluation_metric] for d in list(errors.values())]
        min_err_indices = np.argsort(errs)[:5]
        min_errs = np.array(errs)[min_err_indices]
        best_decomp_names = np.array(list(errors.keys()))[min_err_indices]
        min_errors[dataset] = dict(zip(best_decomp_names,min_errs))

for dataset, bests in min_errors.items():
    print(dataset)
    for k,v in bests.items():
        print(k,v)
    print()
```
